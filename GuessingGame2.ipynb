```python
import numpy as np
import sys

def start() :
    playagain = True
    max = inp()
    guessinggame(max)

    
def inp() :
    x = ''
    while x is not int :
        try : 
            x = int(input("""
Please enter a maximum number of a range : """))
            return x
        except ValueError :
            print("""
            YOU ENTERED A NON-NUMERICAL VALUE !""")


def playagain() :
    repeat = str(input("""
Do you want to play again? (Y/N) : """))
    
    if (repeat.upper() == 'Y') or (repeat.upper() == 'N') :
        if repeat.upper() == 'Y' :
            playagain = True
            max = inp()
            guessinggame(max)
        elif repeat.upper() == 'N' :
            print("""
                        GoodBye !""") 
            sys.exit()
    
    else :
        while (repeat.upper() != 'Y') or (repeat.upper() != 'N') :
            repeat = str(input("""
YOU ENTERED A WRONG CODE! PLEASE ENTER AGAIN! Do you want to play again? (Y/N) : """))

            if repeat.upper() == 'Y' :
                playagain = True
                max = inp()
                guessinggame(max)
            elif repeat.upper() == 'N' :
                print("""
                            GoodBye !""")
                sys.exit()

    

def guessinggame(max) :
   
    while playagain :
        trial_max = 3
        trial = 3
        round = 0
        correct_guess = np.random.randint(0,max,1) 
        
        while round <= trial_max :
            try :
                guess_num = int(input(f"""
You have {trial} chances. Please guess a number between 0 and {max} : """))
                
                if guess_num == correct_guess :
                    print("""
                    YOU GUESSED CORRECTLY ! YOU WIN !!!""")
                    playagain()
                elif (guess_num < correct_guess) and (round != trial_max-1) :
                    print("""
                    The number is too low. Please try again!""")
                    round += 1
                elif (guess_num < correct_guess) and (round == trial_max-1) :
                    print("""
                    The number is too low. YOU LOOSE!""")
                    playagain()
                elif (guess_num > correct_guess) and (round != trial_max-1) :
                    print("""
                    The number is too high. Please try again!""")
                    round += 1
                elif (guess_num > correct_guess) and (round == trial_max-1) :
                    print("""
                    The number is too high. YOU LOOSE!""")
                    playagain()
                trial = trial-1
            
            except ValueError :
                if round == 2 :
                    print("""
                    YOU ENTERED A NON-NUMERICAL VALUE ! YOU LOOSE !!!""")
                    playagain()
                else : 
                    print("""
                    ENTER A NUMBER FOR GUESSING !!!""")
                    round += 1
                    trial = trial-1  


start()
```
