```python
import numpy as np


def inp() :
    x = ''
    while x is not int :
        try : 
            x = int(input("""
Please enter a maximum number of a range : """))
            return x
        except ValueError :
            print("""
            YOU ENTERED A NON-NUMERICAL VALUE !""")

def guessinggame(max) :
    trial_max = 3
    trial = 3
    round = 0
    
    correct_guess = np.random.randint(0,max,1)
    print(correct_guess)

    while round != trial_max :
        try :
            guess_num = int(input(f"""
You have {trial} chances. Please guess a number between 0 and {max} : """))
            if guess_num == correct_guess :
                print("""
                YOU GUESSED CORRECTLY ! YOU WIN !!!""")
                break
            elif (guess_num < correct_guess) and (round != trial_max-1) :
                print("""
                The number is too low. Please try again!""")
                round += 1
            elif (guess_num < correct_guess) and (round == trial_max-1) :
                print("""
                The number is too low. YOU LOOSE!""")
                break
            elif (guess_num > correct_guess) and (round != trial_max-1) :
                print("""
                The number is too high. Please try again!""")
                round += 1
            elif (guess_num > correct_guess) and (round == trial_max-1) :
                print("""
                The number is too high. YOU LOOSE!""")
                break
            trial = trial-1
        except ValueError :
            if round == 2 :
                print("""
                YOU ENTERED A NON-NUMERICAL VALUE ! YOU LOOSE !!!""")
                break
            print("""
            ENTER A NUMBER FOR GUESSING !!!""")
            round += 1
            trial = trial-1

            
max = inp()
guessinggame(max)
```
